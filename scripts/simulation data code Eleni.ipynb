## Signal with bursts (code Elni)


```python
noise = np.random.normal(loc=1, scale=0.3, size=len(time))
```

```python
# create a gaussian function to use as burst signal
gaussian_function = lambda x, mean, sigma: 1/(np.sqrt(2*np.pi*sigma**2))*np.exp(-(x-mean)**2/(2*sigma**2))
```

```python
# create the periodic radio bursts
sigma = 0.01
burst_interval = 0.4 ##in seconds
index_tot = N*(end-begin)
burst_interval_index = (interval/burst_interval)*index_tot
signal_bursts = noise
burst_counter = 0
a = 0
```

```python
for index in range(len(time)):
    bursts_passed = index * interval/burst_interval
    a = a + interval/burst_interval
    #create a check for the if statement to add the burst at the chosen burst interval
    check_burst = bursts_passed/burst_interval
    
    #add every signal to the noise array
    if check_burst == round(check_burst):
        burst_counter = burst_counter + 1
        mean = burst_counter*burst_interval
        burst = [gaussian_function(x=value, mean=mean ,sigma=sigma)/40 for value in time] ## with the number after the gaussian_function you can make the amplitude of the signal smaller
        signal_bursts = signal_bursts + burst
```

```python
# convert from bins to frequency
time_count = len(time)
values = np.arange(int(time_count/2))
time_period = time_count / N
frequencies = values / time_period
```

```python
# do a fourier transformation of simulated data
transform_bursts= np.fft.fft(signal_bursts)/len(signal_bursts)
transform_bursts = transform_bursts[range(int(len(signal_bursts)/2))] # Exclude sampling frequency
```

```python
# plot the simulated noise with burst signal in it
plt.clf()
plt.ylabel('intensity')
plt.xlabel('time (s)')
plt.plot(time, signal_bursts)
```

```python
# plot the fourier transformation of the simulated data
plt.clf()
plt.ylabel('intensity')
plt.xlabel('frequency (Hz)')
plt.plot(frequencies, np.abs(transform_bursts))
plt.ylim(0,0.1) # to crop away the signal at 0
```
